```python
import numpy as np
import matplotlib.pyplot as plt

def I(c):
    '''
    Created definition for the iteration of the function z = z**2 + c. 
    
    Equation that creates the Mandelbrot set. Here, the threshold has been set to 10, meaning that if 
    the abs(z) is greater than thiis value is is assumed it will go to infinity.
    Otherwise, it will be bounded and return to 0 (min could also be se t to -1)
    '''
    z = 0
    for i in range(100):
        if abs(z) > 10:
            return i
        z = z**2 + c
    return 0


# Created two 1D (x and y) arrays and one 2D array (C) that array stores the iteration counts for each point.

X = np.linspace(-2, 2, 200)
Y = np.linspace(-2, 2, 200)
C = np.zeros((200, 200))


for i, x in enumerate(X):
    '''
    Allow us to iterate over each point in complex plane where c is a complex value composed 
    of real (x) and imaginary(y) parts. Here the imaginary value is denoted as j_c
    '''
    for j, y in enumerate(Y):
        j_c = 1j
        c = x + j_c*y
        C[j, i] = I(c)


# Setting the color of the map to 'hot' allows points that diverge to be clearly identified as the brightest shades

plt.imshow(C, cmap='hot')
plt.xlabel('Re(c)')
plt.ylabel('Im(c)') 
plt.show()
```

```python
# Added the color scale to see the values at which the given points diverge

plt.imshow(C)
plt.colorbar()
plt.xlabel('Re(c)')
plt.ylabel('Im(c)') 
plt.show()
```
